# Importing Libraries

```python
import warnings
warnings.filterwarnings("ignore")

import os
import sys
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
sys.path.append('/home/chayan/NeuroMDAVIS/')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utills import Datasets as ds
from sklearn import preprocessing
from utills import NeuroDAVIS as nd
from matplotlib.colors import ListedColormap
```

# Loading Data

```python
x = pd.read_csv('/home/chayan/NeuroMDAVIS/Codev7/Single Omics/Wong/Wong10.csv',index_col=0, header=0)
y = pd.read_csv('/home/chayan/NeuroMDAVIS/Codev7/Single Omics/Wong/Wong10_labels.csv',index_col=0, header=0)
```

```python
idx = pd.DataFrame(preprocessing.LabelEncoder().fit_transform(np.ravel(y['label'])))

col= ['blue', 'green', 'red', 'purple','orange', 'yellow']
classes = list(np.unique(y['label']))
values = list(idx[0])
colors = ListedColormap(col)
```

# Applying NeuroDAVIS

```python
X_embedding = nd.NeuroDAVIS(x, dim=2, lambda_act= 0, lambda_weight= 0, num_neuron=[100, 200],
                          bs=128, epoch=1000, sd=0, verbose=1)

# saving the embedding
pd.DataFrame(X_embedding).to_csv('/home/chayan/NeuroMDAVIS/Codev7/Single Omics/Wong/Wong10_NeuroDAVIS.csv')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(12, 8))
scatter = axes.scatter(X_embedding[:,0],X_embedding[:,1], c=values, cmap=colors, s=1)
axes.set_xlabel('NeuroDAVIS1')
axes.set_ylabel('NeuroDAVIS2')
axes.legend(handles=scatter.legend_elements(num=None)[0], labels=classes, bbox_to_anchor = (0.9, -0.1), ncol=6)
plt.savefig('/home/chayan/NeuroMDAVIS/Codev7/Single Omics/Wong/Wong10_NeuroDAVIS.eps', dpi = 1200, format = 'eps')
```
